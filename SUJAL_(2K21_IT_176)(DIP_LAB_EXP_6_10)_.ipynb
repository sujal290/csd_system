```python
#  EXP 6 :- WAP for blurring an image using Kernel of shape.

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image = cv2.imread('/content/dip.jpg')
kernel_size = (5, 5)
kernel = np.ones(kernel_size, np.float32) / (kernel_size[0] * kernel_size[1])
blurred_image = cv2.filter2D(image, -1, kernel)
cv2_imshow(image)
cv2_imshow(blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('blurred_image.jpg', blurred_image)

```

```python
#  EXP 7:- Prepare a program to find the Contours in an image.

import cv2
import numpy as np
from google.colab.patches import cv2_imshow


image = cv2.imread('/content/dip.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


_, threshold_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)


contours, _ = cv2.findContours(threshold_image, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)


contoured_image = image.copy()
cv2.drawContours(contoured_image, contours, -1, (0, 255, 0), 2)


cv2_imshow(image)
cv2_imshow(threshold_image)
cv2_imshow(contoured_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('contoured_image.jpg', contoured_image)
```

```python
# EXP :- 8 Prepare a program in python to write text on an image.

import cv2
from google.colab.patches import cv2_imshow


image = cv2.imread('/content/dip.jpg')
text = "GSG"
position = (50, 50)
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
color = (255, 0, 0)
thickness = 2
cv2.putText(image, text, position, font, font_scale, color, thickness)

cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('image_with_text.jpg', image)
```

```python
# EXP :- 9 WAP to perform bit plane slicing of a grayscale image.

import numpy as np
import matplotlib.pyplot as plt
import cv2


image = cv2.imread('/content/images.png', cv2.IMREAD_GRAYSCALE)


finalImage = np.zeros(image.shape, dtype=np.uint8)


plt.figure(figsize=(12, 6))

for k in range(8):
    bitplane = np.zeros(image.shape, dtype=np.uint8)
    bitplane[image & (1 << k) != 0] = 255
    finalImage += bitplane * (2 ** k)


    plt.subplot(2, 5, k + 1)
    plt.title(f"BP-{k}")
    plt.imshow(bitplane, cmap='gray')
    plt.axis('off')

plt.subplot(2, 5, 9)
plt.title("Reconstructed")
plt.imshow(finalImage, cmap='gray')
plt.axis('off')

plt.show()
```

```python
# EXP10 :-WAP to display the use of
#          a)3*3 Averaging filter kernal
#          b)5*5 Averaging filter kernal
#          c)10*15 Averaging filter kernal



import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread('/content/dip.jpg')

def apply_averaging_filter(image, kernel_size):

    kernel = np.ones(kernel_size, np.float32) / (kernel_size[0] * kernel_size[1])

    return cv2.filter2D(image, -1, kernel)


image_3x3 = apply_averaging_filter(image, (3, 3))


image_5x5 = apply_averaging_filter(image, (5, 5))


image_10x15 = apply_averaging_filter(image, (10, 15))


plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(image_3x3, cv2.COLOR_BGR2RGB))
plt.title('3x3 Averaging Filter')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(image_5x5, cv2.COLOR_BGR2RGB))
plt.title('5x5 Averaging Filter')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(image_10x15, cv2.COLOR_BGR2RGB))
plt.title('10x15 Averaging Filter')
plt.axis('off')

plt.show()


cv2.imwrite('image_3x3.jpg', image_3x3)
cv2.imwrite('image_5x5.jpg', image_5x5)
cv2.imwrite('image_10x15.jpg', image_10x15)
```
